```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from nltk.corpus import stopwords
import nltk
```

```python
nltk.download('stopwords')
```

```python
df = pd.read_csv('./data/labeled_data.csv')
df.describe()
```

```python
df.head()
```

```python
txt = df.iloc[5913]['tweet']
print(txt)
```

```python
stop_words = set(stopwords.words('english'))
df['tweet']=df['tweet'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stop_words]))
```

```python
words = df['tweet'].str.split().explode().value_counts()
words
```

```python
plt.figure(figsize=(10, 5))
plt.hist(np.log1p(words), bins=30, edgecolor='black', alpha=0.7)
plt.xlabel("Log(Frequency of Words)")
plt.ylabel("Count of Words")
plt.title("Word Frequency Distribution (Log Scale)")
plt.show()
```

```python
top_words = words[:50]  # Only top 50 words
top_words.plot(kind='bar', figsize=(12, 5), color='cornflowerblue', edgecolor='black')
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Top 50 Most Frequent Words")
plt.xticks(rotation=45)
plt.show()
```
